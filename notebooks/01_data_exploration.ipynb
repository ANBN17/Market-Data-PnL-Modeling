# Market Data Analysis & PnL Modeling

This notebook loads equity price data, computes returns, PnL, volatility, and drawdowns.

```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
tickers = ["AAPL", "MSFT", "SPY"]
start = "2020-01-01"
end = "2024-01-01"

data = yf.download(tickers, start=start, end=end, auto_adjust=True)["Close"]
data.tail()
```

```python
returns = data.pct_change().dropna()
cum = (1 + returns).cumprod()

cum.plot(title="Cumulative Returns (Equity Curves)")
plt.show()
```
